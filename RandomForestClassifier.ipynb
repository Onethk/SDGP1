```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from  sklearn.metrics  import accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("DatasetFinal.csv")
df
```

```python
# df = df.drop(["Gender","TravelMethod","TravelTime","PartTimeJob","Relationship","Father's_education","Mother's_education","StudyHours","AttendenceLecs","Concentrate","Friends","SocialMediaTime","extra curricular","HowYouPrepare","StudyPattern","Rate yourself"], axis=1)
x = df.drop("Marks", axis=1)
y = df["Marks"]
```

```python
model.score(x_train,y_train)
```
